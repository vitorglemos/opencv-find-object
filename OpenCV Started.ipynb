```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('pato.jpeg')
```

```python
plt.imshow(img)
```

```python
new = img[50: 500, 50: 500]
```

```python
plt.imshow(new)
```

```python
resize = cv2.resize(img, (200, 200))
plt.imshow(resize)
```

```python
retangle = img 
```

```python
line = cv2.line(retangle, (100, 50),
                  (300, 250),
                  (150, 100, 130), 3)
```

```python
plt.imshow(line)
```

```python
retangle_img = cv2.rectangle(img, (100, 50), (300, 250),
                            (50, 10, 110), 4)

plt.imshow(retangle_img)
```

```python
circle = cv2.circle(img, (100, 50), 100, (200, 80, 150), 5)
```

```python
plt.imshow(circle)
```

```python
text = cv2.putText(img, 'Hello' , (50, 50),
                  cv2.FONT_HERSHEY_SIMPLEX, 2, (200, 200, 200), 2,
                  cv2.LINE_AA)
```

```python
plt.imshow(text)
```
